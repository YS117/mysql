# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
a=random.randint(1000,9999)
print(a)
b=eval(input('please'))
c=1
while b!= a :
    c=c+1
    if a==b:
        print('right')
    if c==4:
        print('QAQ')
        break
    else:
        a=random.randint(1000,9999)
        print(a)
        b=eval(input('please'))       
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

```python
for i in range(10):
    if i==5:
        continue#跳过5
    print(i)
```

```python
for i in range(10):
    for j in range(10):#子循环执行完毕一次之后执行副循环
        for k in range(10):
            if k == 5:
                break
            print(i,j,k)
```

```python
for i in range(1,10):
    for j in range(1,i+1):
        print(i,"x",j,"=",i*j,end='  ')
    print()
```

```python
l =int(input('区间最小值'))
h =int(input('区间最大值'))
for n in range(l,h+1):
    if n > 1:
        for i in range(2,n):
            if(n%i)==0:
                break
            else:
                print(n)
```
